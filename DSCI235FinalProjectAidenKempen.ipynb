## **Advanced NBA Stats**

**Too more accuratly combine stats to evaluate players efficiency we can use advanced stats. The two advanced stats I want to implement into my dataset are PER (Player Efficiency Rating) which has the equation (PTS + REB + AST + STL + BLK âˆ’ Missed FG âˆ’ Missed FT - TO) / GP, and TS% (True Shooting Percentage) which has the equation PTS/(2*(FGA+0.44*FTA)).**

```python
# Player Efficiency Rating (PER)
PTS = df.groupby('Name')['PTS'].sum()
REB = df.groupby('Name')['TRB'].sum()
Merge = pd.merge(PTS, REB, left_index=True, right_index=True)
AST = df.groupby('Name')['AST'].sum()
Merge = pd.merge(Merge, AST, left_index=True, right_index=True)
STL = df.groupby('Name')['STL'].sum()
Merge = pd.merge(Merge, STL, left_index=True, right_index=True)
BLK = df.groupby('Name')['BLK'].sum()
Merge = pd.merge(Merge, BLK, left_index=True, right_index=True)
Merge['MFG'] = df.groupby('Name')['FGA'].sum() - df.groupby('Name')['FG'].sum()
Merge['MFT'] = df.groupby('Name')['FTA'].sum() - df.groupby('Name')['FT'].sum()
Merge['TO'] = df.groupby('Name')['TOV'].sum()
Merge['GP'] = df.groupby('Name')['G'].sum()
Merge
```

**Above I have a dataframe that contains all the stats I need to find the career PER for each player. That stats or columns in order are total career points (PTS), total career rebounds (REB), total career assists (AST), total Career steals (STL), total career blocks (BLK), total career missed field goals (MFG), total career missed free throws (MFT), and total career games played (GP).**

```python
Merge['PER'] = (Merge['PTS']+Merge['TRB']+Merge['AST']+Merge['STL']+Merge['BLK']-Merge['MFG']-Merge['MFT']- Merge['TO'])/(Merge['GP'])
Merge
```

**Above is a data set that has a column for the Player Efficiency Rating (PER) for each player.**

```python
print(Merge['PER'].idxmax())
print(Merge['PER'].max())
```

**It looks like Joel Embiid ended up having the highest Player Effeciency Rating (PER) of 30.24 while Nikola Jokic was second with a PER of 29.39.**

```python
print(Merge['PER'].idxmin())
print(Merge['PER'].min())
```

**It looks like Devin Booker ended up having the lowest Player Effeciency Rating (PER) of 20.10 while Jayson Tatum was second lowest with a PER of 22.22.**

```python
# True Shooting Percentage (TS%)
FGA = df.groupby('Name')['FGA'].sum()
Merge2 = pd.merge(PTS, FGA, left_index=True, right_index=True)
Merge2['FTA'] = df.groupby('Name')['FTA'].sum()
Merge2
```

**Above I have a dataframe that contains all the stats I need to find the career TS% for each player. That stats or columns in order are total career points (PTS), total career field goal attempts (FGA), and total career free throw attempts (FTA).**

```python
Merge2['TS%'] = (Merge2['PTS'])/(2*(Merge2['FGA']+(0.44*Merge2['FTA'])))
Merge2
```

**Above is a data set that has a column for the True Shooting Percentage (TS%) for each player.**

```python
print(Merge2['TS%'].idxmax())
print(Merge2['TS%'].max())
```

**It looks like Nikola Jokic ended up having the highest True Shooting Percentage (TS%) of 0.63 while Joel Embiid was second with a TS% of 0.61.**

```python
print(Merge2['TS%'].idxmin())
print(Merge2['TS%'].min())
```

**It looks like Devin Booker ended up having the lowest True Shooting Percentage (TS%) of 0.57 while Jayson Tatum was second last with a TS% of 0.58.**

**Overall Joel Embiid and Nikola Jokic seem to be the top two players in these advanced NBA stats with Joel being 1st in PER and 2nd in TS%, and Jokic being 1st in TS% and 2nd in PER.**
